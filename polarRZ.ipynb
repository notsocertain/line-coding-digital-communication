```python
import numpy as np
import matplotlib.pyplot as plt

b = [1, 0, 1, 1, 1, 0]
t = np.arange(0, len(b), 0.01)

j = 0
p = []

for i in range(len(t)):
    if t[i] < (j + 0.5):  # In the pulse duration
        if b[j] == 1:
            p.append(1)
        else:
            p.append(-1)  # In polar RZ, binary 0 is represented by a negative pulse
    else:
        p.append(0)  # After the pulse, return to zero
        if t[i] >= (j + 1):  # Move to the next bit
            j += 1

plt.plot(t, p)
plt.xlabel('Time')
plt.ylabel('Voltage')
plt.title('Polar RZ Encoding')
plt.ylim(-1.1, 1.1)  # Set y-axis limits
plt.show()
```
